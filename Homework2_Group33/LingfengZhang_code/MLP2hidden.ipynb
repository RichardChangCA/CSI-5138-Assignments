```
import tensorflow as tf
import matplotlib.pyplot as plt

#import mnist data in tensorflow
from tensorflow.examples.tutorials.mnist import input_data

# from datetime import datetime
# TIMESTAMP = "{0:%Y-%m-%dT%H-%M-%S/}".format(datetime.now())

mnist = input_data.read_data_sets("MNIST_data/",one_hot = True)

def compare_models(use_batch_norm,dropout_value):
  test_accuracy_list = []
  
  #create session
  sess = tf.InteractiveSession()
  
  learning_rate = 0.001

  #input layer number 28*28
  in_units = 784

  #hidden layer number
  h1_units = 512
  h2_units = 256
#   h3_units = 128
#   h3_units = 300
#   h4_units = 300

  #class number
  class_number = 10

  #initialize weights
  W1 = tf.Variable(tf.truncated_normal([in_units,h1_units],mean=0,stddev=0.2))
  b1 = tf.Variable(tf.random_normal([h1_units]))

  W2 = tf.Variable(tf.truncated_normal([h1_units,h2_units],mean=0,stddev=0.2))
  b2 = tf.Variable(tf.random_normal([h2_units]))
  
#   W3 = tf.Variable(tf.truncated_normal([h2_units,h3_units],mean=0,stddev=0.2))
#   b3 = tf.Variable(tf.random_normal([h3_units]))
  
#   W4 = tf.Variable(tf.truncated_normal([h3_units,h4_units],mean=0,stddev=0.2))
#   b4 = tf.Variable(tf.random_normal([h4_units]))

#   W5 = tf.Variable(tf.truncated_normal([h4_units,class_number],mean=0,stddev=0.2))
#   b5 = tf.Variable(tf.random_normal([class_number]))

  W3 = tf.Variable(tf.truncated_normal([h2_units,class_number],mean=0,stddev=0.2))
  b3 = tf.Variable(tf.random_normal([class_number]))

  #define input x
  x = tf.placeholder(tf.float32,[None,in_units])
  #shape must be rank 2

  #define batch norm variable
  is_training = tf.placeholder(tf.bool,name="is_traning")
  # use_batch_norm = tf.placeholder(tf.bool)

  #define dropout probability
  keep_prob = tf.placeholder(tf.float32)

  #layer 1
  hidden1_linear = tf.matmul(x,W1) + b1

#   use_batch_norm= True

  if use_batch_norm:
    hidden1_linear = tf.layers.batch_normalization(hidden1_linear,training=is_training)

  hidden1 = tf.nn.relu(hidden1_linear)

  hidden1_dropout = tf.nn.dropout(hidden1,keep_prob)
  
#   #layer 2
#   hidden2_linear = tf.matmul(hidden1_dropout,W2) + b2

# #   use_batch_norm= True

#   if use_batch_norm:
#     hidden2_linear = tf.layers.batch_normalization(hidden2_linear,training=is_training)

#   hidden2 = tf.nn.relu(hidden2_linear)

#   hidden2_dropout = tf.nn.dropout(hidden2,keep_prob)
  
  
#   #layer 3
#   hidden3_linear = tf.matmul(hidden2_dropout,W3) + b3

# #   use_batch_norm= True

#   if use_batch_norm:
#     hidden3_linear = tf.layers.batch_normalization(hidden3_linear,training=is_training)

#   hidden3 = tf.nn.relu(hidden3_linear)

#   hidden3_dropout = tf.nn.dropout(hidden3,keep_prob)

#   #layer 4
#   hidden4_linear = tf.matmul(hidden3_dropout,W4) + b4
  
#   #do not use batch norm

#   hidden4 = tf.nn.relu(hidden4_linear)

#   #do not use dropout

#   y_ = tf.nn.softmax(tf.matmul(hidden4,W5) + b5)


  hidden2_linear = tf.matmul(hidden1_dropout,W2) + b2
  
  #do not use batch norm

  hidden2 = tf.nn.relu(hidden2_linear)

  #do not use dropout

  y_ = tf.nn.softmax(tf.matmul(hidden2,W3) + b3)

  #define training label
  y = tf.placeholder(tf.float32,[None,class_number])

  #compress to dimension 1
  loss_function = tf.reduce_mean(tf.reduce_sum((y_ - y) ** 2, reduction_indices=[1]))

  if use_batch_norm:
    with tf.control_dependencies(tf.get_collection(tf.GraphKeys.UPDATE_OPS)):
      train_step = tf.train.AdamOptimizer(learning_rate).minimize(loss_function)
  else:
    train_step = tf.train.AdamOptimizer(learning_rate).minimize(loss_function)

  correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
  accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
  # tf.summary.scalar('accuracy', accuracy)

  #activate all defined session part
  tf.global_variables_initializer().run()

  learning_time = 5000

#   test_writer = tf.summary.FileWriter("." + TIMESTAMP, graph=tf.get_default_graph())
  #training without dropout
  for time in range(learning_time):
    batch_xs, batch_ys = mnist.train.next_batch(300)
    train_step.run({x:batch_xs,y:batch_ys,keep_prob:dropout_value,is_training:True})
    if time % 100 == 0:
      learning_rate = learning_rate ** 0.95
      #learning rate decay
    if time % 50 == 0:
      print(time)

      test_accuracy = accuracy.eval({x:mnist.test.images,y:mnist.test.labels,keep_prob:1.0,is_training:False})
#       print(test_accuracy)
      test_accuracy_list.append(test_accuracy)
  #     test_writer.add_summary(test_accuracy, time/100)
  return test_accuracy_list
  

dropout = compare_models(use_batch_norm=False,dropout_value=0.8)
batch_norm = compare_models(use_batch_norm=True,dropout_value=1.0)
common = compare_models(use_batch_norm=False,dropout_value=1.0)

plt.xlabel("epoch")
plt.ylabel("test_accuracy")
plt.ylim((0.9,1))
plt.plot(range(100),dropout,'b-',label="dropout")
plt.plot(range(100),batch_norm,'y-',label="batch_norm")
plt.plot(range(100),common,'r-',label="without")
plt.legend(loc='lower right', fontsize='small')
plt.show()
```
